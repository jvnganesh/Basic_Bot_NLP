```python
import spacy
import gradio as gr
from difflib import get_close_matches

nlp = spacy.load('en_core_web_sm')

responses = {
    "hello": "Hi there! How can I help you today?",
    "hi": "Hello! What can I do for you?",
    "hey": "Hey! How's it going?",
    "how are you": "I'm just a bunch of code, but I'm here to help you!",
    "what's your name": "I'm {bot_name}, your virtual assistant.",
    "who are you": "I'm {bot_name}, a virtual assistant created to help you.",
    "what can you do": "I can chat with you and try to help with basic questions.",
    "help": "Sure, I'm here to help! What do you need assistance with?",
    "what time is it": "I don't have a clock, but you can check the time on your device.",
    "tell me a joke": "Why don't scientists trust atoms? Because they make up everything!",
    "what's the weather like": "I can't check the weather, but you can use a weather app or website for that.\n https://www.accuweather.com/",
    "bye": "Goodbye! Have a great day!",
    "see you later": "See you later! Take care!",
    "goodbye": "Goodbye! It was nice talking to you!",
    "thank you": "You're welcome! If you have any more questions, feel free to ask.",
    "thanks": "No problem! Happy to help.",
    "yes": "Great! How else can I assist you?",
    "no": "Alright, let me know if you need anything else.",
    "maybe": "Okay, let me know if you decide.",
    "what is your purpose": "I'm here to assist you with any questions or tasks I can help with.",
    "can you help me": "Of course! What do you need help with?",
    "where are you from": "I'm a virtual assistant created to help you, so I don't have a physical location.",
    "how old are you": "I'm as old as the code that created me!",
    "what is your favorite color": "I don't have preferences, but I hear blue is quite popular!",
    "do you like music": "I don't have preferences, but music is a great way to relax and enjoy!",
    "what's your favorite movie": "I don't watch movies, but I hear 'The Matrix' is quite interesting!",
    "do you have friends": "I'm a virtual assistant, so I don't have friends, but I'm here to be friendly with you!",
    "what do you like to do": "I like to assist you with your questions and tasks!",
    "do you have hobbies": "My hobby is helping users like you!",
    "are you human": "No, I'm a virtual assistant, here to help you with your questions.",
    "do you have feelings": "I don't have feelings, but I'm here to assist you as best as I can!",
    "what's your favorite food": "I don't eat, but I hear pizza is quite popular!",
    "do you have a family": "I'm a virtual assistant, so I don't have a family, but I'm here to help you!",
    "what is the meaning of life": "That's a big question! Many say it's to find happiness and fulfillment.",
    "what's your favorite book": "I don't read books, but I hear '1984' by George Orwell is quite thought-provoking.",
    "do you like sports": "I don't play sports, but they're a great way to stay active and healthy!",
    "what's your favorite sport": "I don't have a favorite sport, but soccer is popular worldwide!",
    "do you have a pet": "I don't have pets, but pets are wonderful companions!",
    "what's your favorite animal": "I don't have preferences, but many people love dogs for their loyalty.",
    "tell me something interesting": "Did you know that honey never spoils? Archaeologists have found pots of honey in ancient Egyptian tombs that are over 3,000 years old and still edible!",
    "do you know any facts": "Sure! A group of flamingos is called a 'flamboyance'.",
    "what's your favorite place": "I don't have a favorite place, but many people love visiting the beach for relaxation.",
    "can you tell me a story": "Once upon a time, in a land far, far away, there was a chatbot that loved helping people...",
    "what's your favorite song": "I don't listen to music, but 'Bohemian Rhapsody' by Queen is a classic!",
    "do you know any poems": "Sure! 'Roses are red, violets are blue, I'm a chatbot, here to help you!'",
    "can you sing": "I can't sing, but I can try to make you smile with my responses!",
    "what do you think about AI": "AI is a fascinating field that's constantly evolving and helping us in many ways.",
    "do you believe in ghosts": "I don't have beliefs, but ghost stories can be intriguing!",
    "what's your favorite TV show": "I don't watch TV, but many people enjoy 'Friends' for its humor and relatable characters.",
    "what's your favorite season": "I don't experience seasons, but many people love the warmth and activities of summer.",
    "do you like traveling": "I don't travel, but exploring new places can be very exciting and enriching!",
    "do you have any advice": "Always stay curious and keep learning new things!"
}

bot_name = "Chatbot"  # We can put whatever name we want here or in the UI we can give the input for bot name.
def set_bot_name(name):
    global bot_name
    bot_name = name
    return f"Nice to meet you! I'm {bot_name}, your virtual assistant."

def get_response(user_input):
    # List of questions
    questions = list(responses.keys())
    # Find the closest matching question
    closest_match = get_close_matches(user_input.lower(), questions, n=1, cutoff=0.5)
    if closest_match:
        response = responses[closest_match[0]]
    else:
        response = "I'm sorry, I don't understand. Can you please rephrase your question?"
    return response.replace("{bot_name}", bot_name)

def chat(user_input):
    return get_response(user_input)

iface = gr.Blocks()

with iface:
    gr.Markdown("## Welcome to the Chatbot Interface!")

    with gr.Row():
        bot_name_input = gr.Textbox(label="What would you like to name the bot?", placeholder="Enter bot name...")
        bot_name_output = gr.Textbox(label="Bot Introduction")
        bot_name_button = gr.Button("Set Bot Name")

    bot_name_button.click(fn=set_bot_name, inputs=bot_name_input, outputs=bot_name_output)

    with gr.Row():
        chat_input = gr.Textbox(label="You", placeholder="Type your message here...")
        chat_output = gr.Textbox(label=bot_name, interactive=False)

    chat_input.submit(fn=chat, inputs=chat_input, outputs=chat_output)

iface.launch()
```
